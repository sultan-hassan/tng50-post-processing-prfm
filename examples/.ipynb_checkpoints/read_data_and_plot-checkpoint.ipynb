## Plotting all local ISM distributions (figure 2 in Hassan+2024)

```python
fw = 1 #flag to weight or not with SFR
bin_num = 200

line_style    = ['-','--']
xaxis_lims    = np.array([ [0.0,4.0], [-3.0,5.0], [-5.0,3.0], [-2.0,1.0], [-5.0,3.0], [-4.0,2.0]  ])
ext           = 3.0 # extend axis lim by some order of magnitude
fig, ax = plt.subplots(2,3,  figsize=(13,7))

ind = 0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')

    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])*1e-6 # convert from Msun kpc^2 to Msun pc^2
    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])*1e-6 # convert from Msun kpc^2 to Msun pc^2
    rho_star   = np.hstack([np.hstack(x) for x in df['rho_*']])*1e-9   # convert from Msun kpc^3 to Msun pc^3
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])*1e-9  # convert from Msun kpc^3 to Msun pc^3
    
    generate_kde(np.log10(Sigma_g)    ,Sigma_SFR, ax[0][0], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[0,0]-ext,xaxis_lims[0,1]+ext)
    generate_kde(np.log10(H_star)     ,Sigma_SFR, ax[1][0], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[1,0]-ext,xaxis_lims[1,1]+ext)
    generate_kde(np.log10(Sigma_star) ,Sigma_SFR, ax[0][1], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[2,0]-ext,xaxis_lims[2,1]+ext)
    generate_kde(np.log10(rho_star)   ,Sigma_SFR, ax[1][1], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[4,0]-ext,xaxis_lims[4,1]+ext)
    generate_kde(np.log10(rho_dm)     ,Sigma_SFR, ax[1][2], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[5,0]-ext,xaxis_lims[5,1]+ext)
    generate_kde(np.log10(Sigma_SFR)  ,Sigma_SFR, ax[0][2], bin_num, r'$z=$'+str(i),'k', line_style[ind] ,fw, xaxis_lims[3,0]-ext,xaxis_lims[3,1]+ext)
    
    ind += 1

label_axis  = [ r'$\log_{10}\, \Sigma_{g}\, [M_{\odot}\, {\rm pc^{-2}}]$', r'$\log_{10}\, \Sigma_{\star}\, [M_{\odot}\, {\rm pc^{-2}}]$', r'$\log_{10} \Sigma_{\rm SFR}\, [M_{\odot}\, {\rm yr^{-1}\, kpc^{-2}}]$', r'$\log_{10}\,  H_{\star}\, [{\rm kpc}]$', r'$\log_{10}\,  \rho_{\star}\, [M_{\odot}\, {\rm pc^{-3}}]$', r'$\log_{10}\,  \rho_{d}\, [M_{\odot}\, {\rm pc^{-3}}]$']
xaxis_ticks = [np.arange(xaxis_lims[0,0],xaxis_lims[0,1]+1,1), np.arange(xaxis_lims[1,0],xaxis_lims[1,1]+2,2),np.arange(xaxis_lims[2,0],xaxis_lims[2,1]+2,2), np.arange(xaxis_lims[3,0],xaxis_lims[3,1]+1,1),
               np.arange(xaxis_lims[4,0],xaxis_lims[4,1]+2,2), np.arange(xaxis_lims[5,0],xaxis_lims[5,1]+2,2)]



ind=0
for j in range(2):
    for k in range(3):
        ax[j,k].tick_params(labelsize=18)  
        ax[j,k].set_xlabel(label_axis[ind], fontsize=15)
        ax[j,k].set_xlim([xaxis_lims[ind][0],xaxis_lims[ind][1]])
        ax[j,k].set_xticks(xaxis_ticks[ind])
        ax[j,k].set_ylim(bottom=0)
        ind+=1

ax[0][0].legend(loc='upper right', fancybox=True,shadow=True, fontsize=14)
plt.subplots_adjust(wspace=0.2, hspace=0.3)
fig.supylabel(r'$ \frac{1}{\rm SFR_{\rm tot}}\frac{\rm d\, SFR}{{\rm d\, log_{10}\,} x} $', fontsize=25, y=0.5, x=0.05)
fig.show()
```

## Comparison between various components of pressure and weights (figure 4 in Hassan+2024)

```python
fig, ax = plt.subplots(2,3,sharey=True, sharex='col',  figsize=(12,6))

ind=0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')

    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    # pressue components
    p_th       = np.hstack([np.hstack(x) for x in df['P_th']])
    p_mag      = np.hstack([np.hstack(x) for x in df['P_mag']])
    p_turb     = np.hstack([np.hstack(x) for x in df['P_turb']])
    p_tot      = p_th + p_mag + p_turb
    
    # compute scale height from the simulation pressure.
    rho_g      = np.hstack([np.hstack(x) for x in df['rho_g']])
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])
    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])  
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])
    
    sigma_eff  = np.sqrt( p_tot/rho_g * (kB * pc_per_cm*pc_per_cm*pc_per_cm/10.0/Msun_per_gram)) # km/s      
    A_coeff    = (8.0/3.0) * rho_dm / Sigma_g  # kpc^-1                                                  
    B_coeff    = (1.0 +  2.0*Sigma_star/Sigma_g +  A_coeff*H_star) # unitless  
    C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
    D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
    Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in kpc
    
    # weight components

    W_gas  = np.pi * G * Sigma_g * Sigma_g / 2.0 #    (M⊙ /kpc^3) (km/s)2
    W_gas  *= (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # convert to K/cm^3
    W_star = np.pi * G * Sigma_g * Sigma_star * Hg_equil/ ( Hg_equil + H_star ) # (M⊙ /kpc^3) (km/s)2
    W_star *= (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # convert to K/cm^3
    W_dm   = (4.0/3.0) * np.pi * G * rho_dm  * Sigma_g * Hg_equil # (M⊙ /kpc^3) (km/s)2
    W_dm   *= (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # convert to K/cm^3

    W_tot = W_gas + W_star + W_dm
    
    
    generate_kde(np.log10(p_th/p_tot),Sigma_SFR,   ax[ind][0], bin_num, r'$P_{\rm th}/P_{\rm tot}$',   'k', 'solid' ,fw,-3,0)
    generate_kde(np.log10(p_turb/p_tot),Sigma_SFR, ax[ind][0], bin_num, r'$P_{\rm turb}/P_{\rm tot}$', 'k', 'dashed',fw,-3,0)
    generate_kde(np.log10(p_mag/p_tot),Sigma_SFR,  ax[ind][0], bin_num, r'$\Pi_{\rm mag}/P_{\rm tot}$','k', 'dotted',fw,-3,0)
    
    generate_kde(np.log10(W_gas/W_tot),Sigma_SFR,  ax[ind][1], bin_num, r'$W_{\rm g}/{\cal W}$',       'k', 'solid' ,fw,-3,0)
    generate_kde(np.log10(W_dm/W_tot),Sigma_SFR,   ax[ind][1], bin_num, r'$W_{\rm d}/{\cal W}$',       'k', 'dashed',fw,-3,0)
    generate_kde(np.log10(W_star/W_tot),Sigma_SFR, ax[ind][1], bin_num, r'$W_{\rm \star}/{\cal W}$',   'k', 'dotted',fw,-3,0)

    generate_kde(np.log10(p_tot/W_tot),Sigma_SFR, ax[ind][2], bin_num, None,'k', '-',fw,-2,2)
    ind+=1

    
for j in range(2):
    ax[j,0].set_xlim([-3,0])
    ax[j,1].set_xlim([-2,0])
    ax[j,2].set_xlim([-1,1])  
    ax[0,j].legend(loc='upper left', fontsize=16)
    ax[j,2].axvline(x=0, lw=2, color='k', ls='--')
    for k in range(2):                                                                                                  
        ax[j,k].tick_params(labelsize=18) 
        ax[j,k].set_ylim(bottom=0.0)
       
ax[1,0].set_xlabel(r"$\log_{10}\, P_{i}/P_{\rm tot}$", fontsize=16)
ax[1,1].set_xlabel(r"$\log_{10}\, W_{i}/{\cal W}$", fontsize=16)
ax[1,2].set_xlabel(r"$\log_{10}\, P_{\rm tot}/{\cal W}$", fontsize=16)


fig.supylabel(r'$ \frac{1}{\rm SFR_{\rm tot}}\frac{\rm d\, SFR}{{\rm d\, log_{10}\,} x} $', fontsize=25, y=0.5, x=0.05)
fig.show()
```

## Comparison between various scale heights (figure 6 in Hassan+2024)

```python
fig, ax = plt.subplots(2,2,sharey=True, sharex='col',  figsize=(10,6))


ind = 0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')
    
    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    rho_g      = np.hstack([np.hstack(x) for x in df['rho_g']])
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])
    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])  
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])
    H_g        = np.hstack([np.hstack(x) for x in df['H_g']])
 
    # ===========================
    # using constant velocity
    # ===========================
    
    sigma_eff  = 15.0 # km/s      
    A_coeff    = (8.0/3.0) * rho_dm / Sigma_g  # kpc^-1                                                  
    B_coeff    = (1.0 +  2.0*Sigma_star/Sigma_g +  A_coeff*H_star) # unitless  
    
    
    C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
    D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
    Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in kpc
    
    generate_kde(np.log10(Hg_equil),      Sigma_SFR, ax[ind][0], bin_num, r'$H_{g}(equil., \sigma_{\rm eff}= 15\,{\rm km/s})$','b', '-',fw, -3,3)
    generate_kde(np.log10(H_g/Hg_equil),  Sigma_SFR, ax[ind][1], bin_num, r'$H_{g}(measured)/H_{g}(equil.)$'                  ,'b', '-',fw, -3,3)
    
    
    
    
    # ===========================
    # using sigma_eff from PRFM/TIGRESS fit sigma_eff = ~ W^0.22, we need to solve iteratively 
    # for scale heigh, pressure and velocity till convergence in sigma_eff is achieved 
    # (see a flowchart/figure 3 in Hassan+2024) 
    # convegernce happens fast after 4-5 iterations, lets use 10
    # ===========================
    
    for tt in range(10):
        C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
        D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
        Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in  kpc                                                                                                                                                            
        P_midplane = Sigma_g*sigma_eff*sigma_eff / (2.0 * Hg_equil) * (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # K/ cm^3                                                                                             
        sigma_eff  =  12.0 * ( P_midplane * 1e-4 )**0.22 # from TIGRESS simulations 

    generate_kde(np.log10(Hg_equil),     Sigma_SFR, ax[ind][0], bin_num, r'$H_{g}(equil., \sigma_{\rm eff}\sim {\cal W}^{0.22})$','r', '-',fw, -3,3)
    generate_kde(np.log10(H_g/Hg_equil), Sigma_SFR, ax[ind][1], bin_num, r'$H_{g}(measured)/H_{g}(equil.)$',                      'r', '-',fw, -3,3)
    
    # ===========================
    # using sigma_eff from the simulation pressure sigma_eff = (total pressure/density)^0.5
    # ===========================

    p_th       = np.hstack([np.hstack(x) for x in df['P_th']])
    p_mag      = np.hstack([np.hstack(x) for x in df['P_mag']])
    p_turb     = np.hstack([np.hstack(x) for x in df['P_turb']])
    p_tot      = p_th + p_mag + p_turb
    
    sigma_eff  = np.sqrt( p_tot/rho_g * (kB * pc_per_cm*pc_per_cm*pc_per_cm/10.0/Msun_per_gram)) # km/s 
    C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
    D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
    Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in kpc
    
    generate_kde(np.log10(Hg_equil),     Sigma_SFR, ax[ind][0], bin_num, r'$H_{g}(equil., \sigma_{\rm eff}=\sqrt{P_{\rm tot}/\rho})$','g', '-',fw,-3,3)
    generate_kde(np.log10(H_g/Hg_equil), Sigma_SFR, ax[ind][1], bin_num, r'$H_{g}(measured)/H_{g}(equil.)$',                          'g', '-',fw,-3,3)
    
    
    # ===========================
    # using measured pressure from the simulations
    # ===========================
    generate_kde(np.log10(H_g),  Sigma_SFR, ax[ind][0], bin_num, r'$H_{g}(measured)$','k', '-',fw, -3,3)
    ax[ind][0].annotate(r'$z='+str(i)+'$',xy=(0.05,0.2),xycoords='axes fraction',bbox=dict(boxstyle="round",fc="w"), size=14)
    
    ind +=1 
    

for j in range(2):
    ax[j,1].axvline(x=0.0, ls='--', color='k', lw=2)
    for k in range(2):
        ax[j,0].set_xlim([-3,1])
        ax[j,1].set_xlim([-1,2])                                                                                                                                                                               
        ax[j,k].tick_params(labelsize=18)                                                                                                                                                  
        ax[j,k].set_ylim(bottom=0.0)

ax[1,0].set_xlabel(r"$\log_{10}\, H_{g}\, [\rm kpc]$", fontsize=18)
ax[1,1].set_xlabel(r"$\log_{10}\, H_{g}\, (measured)/ H_{g}(equil.)$", fontsize=18)
ax[0,0].legend(loc='upper left', fontsize=11,frameon=False)

fig.supylabel(r'$ \frac{1}{\rm SFR_{\rm tot}}\frac{\rm d\, SFR}{{\rm d\, log_{10}\,} x} $', fontsize=25, y=0.5, x=0.01)
plt.tight_layout()
fig.show()
```

## Comparison between depletion times (figure 8 in Hassan+2024)

```python
fig, ax = plt.subplots(1,2, sharey=True, sharex='col',figsize=(8,3))

ind = 0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')
    
    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    rho_g      = np.hstack([np.hstack(x) for x in df['rho_g']])
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])
    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])  
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])

    
    sigma_eff  = 15.0 # km/s      
    A_coeff    = (8.0/3.0) * rho_dm / Sigma_g  # kpc^-1                                                  
    B_coeff    = (1.0 +  2.0*Sigma_star/Sigma_g +  A_coeff*H_star) # unitless  
    for tt in range(10):
        C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
        D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
        Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in  kpc                                                                                                                                                            
        t_dyn   = (2.0 * Hg_equil*(pc_per_cm/100.0) / sigma_eff ) * (1.0/Year_per_sec) * 1e-9 ## dynamical time in Gyr
        P_midplane = Sigma_g*sigma_eff*sigma_eff / (2.0 * Hg_equil) * (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # K/ cm^3                                                                                             
        sigma_eff  =  12.0 * ( P_midplane * 1e-4 )**0.22 # from TIGRESS simulations 
        
    Upsilon        = 10.0**(  -0.212*np.log10( P_midplane  )  + 3.86 ) # from TIGRESS simulations
    
    t_dep_TNG50      = Sigma_g/Sigma_SFR  * 1e-9 # depletion time from TNG50 in Gyr                                                                                                                                                   
    t_dep_PRFM       = Upsilon * t_dyn / sigma_eff # depletion time from PRFM in Gyr   
        
    generate_kde(np.log10(t_dep_TNG50), Sigma_SFR, ax[ind], bin_num,  r'$\rm SH03$','k', '-',fw, -4,4)
    generate_kde(np.log10(t_dep_PRFM),  Sigma_SFR, ax[ind], bin_num, r'$\rm PRFM$','k', '--',fw, -4,4)
    
    ax[ind].annotate(r'$z='+str(i)+'$',xy=(0.05,0.3),xycoords='axes fraction',bbox=dict(boxstyle="round",fc="w"), size=15)
    
    ind+=1
    
for i in range(2):                                                                                                                                                         
        ax[i].set_xlim([-3.5,2.5])                                                                                                                                                                 
        ax[i].tick_params(labelsize=16)                                                                                                                                  
        ax[i].set_xticks([-3,-2,-1,0,1,2])                                                                                                                    
        ax[i].set_ylim(bottom=0.0)
        ax[i].set_xlabel(r"$\log_{10}\, t_{\rm dep}\, [\rm Gyr]$", fontsize=18)

ax[0].legend(loc='upper left', fancybox=True,shadow=True, fontsize=12)

fig.supylabel(r'$ \frac{1}{\rm SFR_{\rm tot}}\frac{\rm d\, SFR}{{\rm d\, log_{10}\,} t_{\rm dep}} $', fontsize=20, y=0.6, x=0.01)
plt.tight_layout()
fig.show()
```

## Comparison between depletion times vs dynamical times (figure 9 in Hassan+2024)

```python
fig, ax = plt.subplots(2,2, sharey=True, sharex=True,figsize=(10,10))

ind = 0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')
    
    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    rho_g      = np.hstack([np.hstack(x) for x in df['rho_g']])
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])
    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])  
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])

    
    sigma_eff  = 15.0 # km/s      
    A_coeff    = (8.0/3.0) * rho_dm / Sigma_g  # kpc^-1                                                  
    B_coeff    = (1.0 +  2.0*Sigma_star/Sigma_g +  A_coeff*H_star) # unitless  
    for tt in range(10):
        C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
        D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
        Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in  kpc                                                                                                                                                            
        t_dyn      = (2.0 * Hg_equil*(pc_per_cm/100.0) / sigma_eff ) * (1.0/Year_per_sec) * 1e-9 ## dynamical time in Gyr
        P_midplane = Sigma_g*sigma_eff*sigma_eff / (2.0 * Hg_equil) * (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # K/ cm^3                                                                                             
        sigma_eff  =  12.0 * ( P_midplane * 1e-4 )**0.22 # from TIGRESS simulations 
        
    Upsilon        = 10.0**(  -0.212*np.log10( P_midplane  )  + 3.86 ) # from TIGRESS simulations
    
    t_dep_TNG50      = Sigma_g/Sigma_SFR  * 1e-9 # depletion time from TNG50 in Gyr                                                                                                                                                   
    t_dep_PRFM       = Upsilon * t_dyn / sigma_eff # depletion time from PRFM in Gyr   
        
    im = do_hist2d(np.log10(t_dyn), np.log10(t_dep_PRFM), Sigma_SFR, ax[ind][0], bin_width, 'auto', -5.,5., -5., 5.,"hot_r")
    im = do_hist2d(np.log10(t_dyn), np.log10(t_dep_TNG50), Sigma_SFR, ax[ind][1], bin_width, 'auto', -5.,5., -5., 5.,"hot_r")
    
    
    ax[ind][0].annotate(r'$z='+str(i)+'$',xy=(0.8,0.1),xycoords='axes fraction',bbox=dict(boxstyle="round",fc="w"), size=20)
    
    ind+=1
    
    

# generate some tdyn values for some line reference    
x_tdyn = 10**np.linspace(-4,4, 100)

for j in range(2):
    for k in range(2):
        ax[j,k].set_xlim([-4,0])
        ax[j,k].set_ylim([-3,2])
        ax[j,k].grid(True)
        ax[j,k].tick_params(labelsize=18)
        ax[j][k].plot(np.log10(x_tdyn), np.log10(x_tdyn*1.0),     'g--', label=r'$t_{\rm dep}/t_{\rm dyn}=1$')
        ax[j][k].plot(np.log10(x_tdyn), np.log10(x_tdyn*10.0),    'b--', label=r'$t_{\rm dep}/t_{\rm dyn}=10$')
        ax[j][k].plot(np.log10(x_tdyn), np.log10(x_tdyn*(1/0.02)),'k-' , label=r'$t_{\rm dep}/t_{\rm dyn}=50$')
        ax[j][k].plot(np.log10(x_tdyn), np.log10(x_tdyn*100.0),   'r--', label=r'$t_{\rm dep}/t_{\rm dyn}=100$')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([1.01, 0.18, 0.03, 0.7])
fig.colorbar(im, cax=cbar_ax, extend='min')
cbar_ax.tick_params(labelsize=18)
cbar_ax.set_ylabel(r"${\rm Fraction\,\, of\,\, the\,\, total\,\,  SFR}\,$", rotation=270,fontsize=24, labelpad=30)

for i in range(2):
    ax[i,0].set_ylabel(r"$\log_{10}\, t_{\rm dep}\, [\rm Gyr]$", fontsize=24)
    ax[1,i].set_xlabel(r'$\log_{10}\,  t_{\rm dyn}\, [\rm Gyr]$', fontsize=24)
                                                                                                                                            
ax[0,0].set_title(r"$\rm PRFM\,$", fontsize=26)
ax[0,1].set_title(r"$\rm SH03\,$", fontsize=26)

ax[0][0].legend(loc='upper left', fancybox=True,shadow=True, fontsize=15)                                                                                                                                              
plt.tight_layout()
fig.show()

                   
```

## Comparison between global Schmidt relations (figure 11 in Hassan+2024)

```python
# TIGRESS data from Ostriker & Kim 2022 (arXiv:2206.00681), see Table 2

TIGRESS_Sigma_g  = np.array([70.9,30.2,9.87,2.45,65.9,39.6,11.2])
TIGRESS_SFR      = np.array([1.10,5.37e-2,2.67e-3,6.21e-5,1.17e-1,5.41e-2,2.16e-3])

fig, ax = plt.subplots(2,2, sharey=True, sharex=True,figsize=(10,10))

ind = 0
for i in [0,2]:
    df = pd.read_pickle('../data/all_data_z'+str(i)+'.pkl')
    
    Sigma_SFR  = np.hstack([np.hstack(x) for x in df['Sigma_SFR']])
    rho_g      = np.hstack([np.hstack(x) for x in df['rho_g']])
    rho_dm     = np.hstack([np.hstack(x) for x in df['rho_dm']])
    Sigma_g    = np.hstack([np.hstack(x) for x in df['Sigma_g']])
    Sigma_star = np.hstack([np.hstack(x) for x in df['Sigma_*']])  
    H_star     = np.hstack([np.hstack(x) for x in df['H_*']])

    
    sigma_eff  = 15.0 # km/s      
    A_coeff    = (8.0/3.0) * rho_dm / Sigma_g  # kpc^-1                                                  
    B_coeff    = (1.0 +  2.0*Sigma_star/Sigma_g +  A_coeff*H_star) # unitless  
    for tt in range(10):
        C_coeff    = H_star - sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc    
        D_coeff    = - H_star * sigma_eff*sigma_eff / ( np.pi * G * Sigma_g   ) # kpc^2  
        Hg_equil   = solve_cubic(B_coeff/A_coeff, C_coeff/A_coeff, D_coeff/A_coeff) # scale height from vertical equalibrium in  kpc                                                                                                                                                            
        t_dyn      = (2.0 * Hg_equil*(pc_per_cm/100.0) / sigma_eff ) * (1.0/Year_per_sec) * 1e-9 ## dynamical time in Gyr
        P_midplane = Sigma_g*sigma_eff*sigma_eff / (2.0 * Hg_equil) * (Msun_per_gram / (kB *pc_per_cm *pc_per_cm *pc_per_cm/10)) # K/ cm^3                                                                                             
        sigma_eff  =  12.0 * ( P_midplane * 1e-4 )**0.22 # from TIGRESS simulations 
        
    Upsilon        = 10.0**(  -0.212*np.log10( P_midplane  )  + 3.86 ) # from TIGRESS simulations
    
    
    Sigma_SFR_PRFM = P_midplane/Upsilon * 2.079107930411966e-4  # see Eq 28 in Ostriker & Kim 2022 (arXiv:2206.00681)

    
    im = do_hist2d(np.log10(Sigma_g)-6.0, np.log10(Sigma_SFR_PRFM), Sigma_SFR, ax[ind][0], bin_width, 'auto', -5.,5., -7., 3.,"hot_r")
    
    im = do_hist2d(np.log10(Sigma_g)-6.0, np.log10(Sigma_SFR), Sigma_SFR,      ax[ind][1], bin_width, 'auto', -5.,5., -7., 3.,"hot_r")
    
    
    
    ax[ind][0].plot(np.log10(TIGRESS_Sigma_g), np.log10(TIGRESS_SFR), '*',color='b', ms=12, label=r"$\rm TIGRESS$")
    ax[ind][1].plot(np.log10(TIGRESS_Sigma_g), np.log10(TIGRESS_SFR), '*',color='b', ms=12, label=r"$\rm TIGRESS$")
    ax[ind][0].annotate(r'$z='+str(i)+'$',xy=(0.8,0.05),xycoords='axes fraction',bbox=dict(boxstyle="round",fc="w"), size=20)
    
    ind+=1
    

# generate some Sigma_g values for some line reference

x_Sigma_g = 10**np.linspace(-4,4, 100)

for j in range(2):
    for k in range(2):
        ax[j,k].set_xlim([0.0,3.0])
        ax[j,k].set_ylim([-5,3])
        ax[j,k].set_xticks([0,0.5,1,1.5,2,2.5,3])
        ax[j,k].set_yticks([-5,-4,-3,-2,-1,0,1,2,3])                                                                                                                                                        
        ax[j,k].grid(True)
        ax[j,k].tick_params(labelsize=18)

        #===== log Sigma_SFR = 1.5 log Sigma_gas - 3.87 from Kennicutt & De Los Reyes 2021                                                                                                                                    
        ax[j][k].plot(np.log10(x_Sigma_g), 1.5*np.log10(x_Sigma_g) - 3.87, 'k-',lw=2, label=r'$\rm Fixed\,\, MW\,\, \alpha_{\rm CO} $')                                                  

        ax[j][k].fill_between(np.log10(x_Sigma_g), 1.5*np.log10(x_Sigma_g) - 3.87 - 2*0.4, 1.5*np.log10(x_Sigma_g) - 3.87 + 2*0.4, color='k', alpha=0.2)

        #===== log Sigma_SFR = 1.92 log Sigma_gas - 4.29 from Bolatto+ 2013 alpha_CO                                                                                                                
        ax[j][k].plot(np.log10(x_Sigma_g), 1.92*np.log10(x_Sigma_g) - 4.29, 'k--', lw=2, label=r'$\rm Bolatto+2013\, \alpha_{\rm CO} $')                       
        ax[j][k].fill_between(np.log10(x_Sigma_g), 1.92*np.log10(x_Sigma_g) - 4.29 - 2*0.4, 1.92*np.log10(x_Sigma_g) - 4.29+ 2*0.4, color='k', alpha=0.075)
        
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([1.01, 0.18, 0.03, 0.7])
fig.colorbar(im, cax=cbar_ax, extend='min')
cbar_ax.tick_params(labelsize=18)
cbar_ax.set_ylabel(r"${\rm Fraction\,\, of\,\, the\,\, total\,\,  SFR}\,$", rotation=270,fontsize=24, labelpad=30)
        
for i in range(2):
    ax[i,0].set_ylabel(r"$\log_{10}\, \Sigma_{\rm SFR\,}\, [\rm M_{\odot}/yr/kpc^{2}]$", fontsize=24)
    ax[1,i].set_xlabel(r'$\log_{10}\, \Sigma_{\rm g}\, [\rm M_{\odot}\, pc^{-2}]$', fontsize=24)
    
ax[0,0].set_title(r"$\rm PRFM\,$", fontsize=26)
ax[0,1].set_title(r"$\rm SH03\,$", fontsize=26)

ax[0,1].legend(loc='upper left', fancybox=True,shadow=True, fontsize=16)                                                                                                                                               
plt.tight_layout()
fig.show()
```
